```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
%matplotlib inline
```

```python
data = pd.read_csv('Salary_dataset.csv')
data    
```

```python
# scatter plot
plt.scatter(data['YearsExperience'], data['Salary'])
plt.xlabel('Experience in years')
plt.ylabel('Salary')
plt.show()
```

```python
# correlation 
data.corr()
```

```python
# seaborn for visualization 
import seaborn as sns
```

```python
sns.pairplot(data)
plt.show()
```

```python
# Independent and dependent feature 
X = data[['YearsExperience']]      # your independt feature should be in dataframe(two brackets) or 2d. This is becuase when you train your mdoel you will getter error
Y = data['Salary']  # this is can be in series or 1d array
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.25,random_state=42) 
```

```python
# standardization
from sklearn.preprocessing import StandardScaler 
```

```python
scalar = StandardScaler()
X_train = scalar.fit_transform(X_train)
X_test = scalar.transform(X_test)
```

```python
## Apply simple linear regression  
from sklearn.linear_model import LinearRegression 
regression = LinearRegression()
```

```python
regression.fit(X_train, Y_train)
```

```python
print('coefficient or slope',regression.coef_)
print('interncept',regression.intercept_)
```

```python
#plotting training data along with best fit line
plt.scatter(X_train,Y_train)
plt.plot(X_train,regression.predict(X_train))
plt.show()
```

# Prediction of test data 
1. predicted salary (output) = intercept + coef_(years of experience)
2. y_pred_test = 25613.47 + 74585.59

```python
# prediction for test data 
y_pred = regression.predict(X_test)
```

```python
# performance matrix
from sklearn.metrics import mean_absolute_error,mean_squared_error
```

```python
mae = mean_absolute_error(Y_test,y_pred)
mse = mean_squared_error(Y_test, y_pred)
rmse = np.sqrt(mse)
print(mae), print(mse), print(rmse)
```

# R square
Formula

R^2 = 1 - SSR/SST

R^2 = coefficient of determination 
SSR = sum of squares of residuals 
SST = total sum of squares 

```python
from sklearn.metrics import r2_score
```

```python
score = r2_score(Y_test,y_pred)
score
```
